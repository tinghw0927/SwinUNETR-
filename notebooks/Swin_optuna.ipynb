# ***2.ä¸‹è¼‰è¨“ç·´è³‡æ–™é›†***

```python
# ç’°å¢ƒè¨­å®šèˆ‡è³‡æ–™æº–å‚™

# å®‰è£ PyTorch
!pip install torch==2.1.0 torchvision==0.16.0 torchaudio==2.1.0 --extra-index-url https://download.pytorch.org/whl/cu118

# Import æ‰€æœ‰éœ€è¦çš„å¥—ä»¶
from google.colab import drive
import os
import shutil
import json
import glob
from sklearn.model_selection import train_test_split

# æ›è¼‰ Google Drive
drive.mount('/content/drive')

# è¨­å®šè·¯å¾‘
zip_path = "/content/drive/MyDrive/training_label.zip"
base_path = "/content/CardiacSegV2/dataset/chgh"
os.makedirs(base_path, exist_ok=True)

# è§£å£“ç¸®
if not os.path.exists(zip_path):
    raise FileNotFoundError(f"æ‰¾ä¸åˆ° ZIP æª”æ¡ˆï¼š{zip_path}")

print("é–‹å§‹è§£å£“ç¸®...")
!unzip -q -o "{zip_path}" -d "{base_path}"
print("è§£å£“ç¸®å®Œæˆ")

# æ•´ç†è³‡æ–™å¤¾çµæ§‹
target_image_dir = f"{base_path}/training_image"
target_label_dir = f"{base_path}/training_label"

# æª¢æŸ¥æ˜¯å¦éœ€è¦æ•´ç†
need_reorganize = (
    not os.path.exists(target_image_dir) or
    not os.path.exists(target_label_dir) or
    len([f for f in os.listdir(target_image_dir) if f.endswith('.nii.gz')]) == 0 if os.path.exists(target_image_dir) else True
)

if need_reorganize:
    print("\næ•´ç†æª”æ¡ˆçµæ§‹...")

    # æ‰¾å‡ºè³‡æ–™ä¾†æº
    if os.path.exists(f"{base_path}/training_label/41_training_image_01"):
        source_base = f"{base_path}/training_label"
    elif os.path.exists(f"{base_path}/41_training_image_01"):
        source_base = base_path
    else:
        raise FileNotFoundError("æ‰¾ä¸åˆ°è³‡æ–™ä¾†æºè³‡æ–™å¤¾!")

    # æ¸…ç†ä¸¦å»ºç«‹ç›®æ¨™è³‡æ–™å¤¾
    for dir_path in [target_image_dir, f"{base_path}/training_label_new"]:
        if os.path.exists(dir_path):
            shutil.rmtree(dir_path)
        os.makedirs(dir_path, exist_ok=True)

    # æ•´åˆå½±åƒæª”æ¡ˆï¼ˆå¾ 3 å€‹è³‡æ–™å¤¾ï¼‰
    moved_images = 0
    for i in range(1, 4):
        src_dir = f"{source_base}/41_training_image_0{i}"
        if os.path.exists(src_dir):
            for f in os.listdir(src_dir):
                if f.endswith('.nii.gz'):
                    shutil.move(os.path.join(src_dir, f), target_image_dir)
                    moved_images += 1
            shutil.rmtree(src_dir)

    # æ•´åˆæ¨™ç±¤æª”æ¡ˆ
    src_label_dir = f"{source_base}/41_training_label"
    moved_labels = 0
    if os.path.exists(src_label_dir):
        for f in os.listdir(src_label_dir):
            if f.endswith('.nii.gz'):
                shutil.move(os.path.join(src_label_dir, f), f"{base_path}/training_label_new")
                moved_labels += 1
        shutil.rmtree(src_label_dir)

    # é‡æ–°å‘½åæ¨™ç±¤è³‡æ–™å¤¾
    if os.path.exists(target_label_dir):
        shutil.rmtree(target_label_dir)
    os.rename(f"{base_path}/training_label_new", target_label_dir)

    # æ¸…ç† __MACOSX
    if os.path.exists(f"{base_path}/__MACOSX"):
        shutil.rmtree(f"{base_path}/__MACOSX")

    print(f"æ•´ç†å®Œæˆï¼šå½±åƒ {moved_images} å€‹ï¼Œæ¨™ç±¤ {moved_labels} å€‹")
else:
    print("è³‡æ–™çµæ§‹å·²å°±ç·’ï¼Œè·³éæ•´ç†")

# é…å°å½±åƒå’Œæ¨™ç±¤
print("\nğŸ” æƒæä¸¦é…å°æª”æ¡ˆ...")

images = sorted(glob.glob(f"{base_path}/training_image/*.nii.gz"))
labels = sorted(glob.glob(f"{base_path}/training_label/*.nii.gz"))

dataset = []

# _gt å¾Œç¶´é…å°
for img in images:
    img_name = os.path.basename(img).replace('.nii.gz', '')
    label_path = os.path.join(target_label_dir, f"{img_name}_gt.nii.gz")

    if os.path.exists(label_path):
        dataset.append({"image": img, "label": label_path})

# å¦‚æœå¤±æ•—ï¼Œç›´æ¥åç¨±é…å°
if len(dataset) == 0:
    for img in images:
        label_path = os.path.join(target_label_dir, os.path.basename(img))
        if os.path.exists(label_path):
            dataset.append({"image": img, "label": label_path})

if len(dataset) == 0:
    raise ValueError(f"ç„¡æ³•é…å°ä»»ä½•æª”æ¡ˆï¼å½±åƒï¼š{len(images)}ï¼Œæ¨™ç±¤ï¼š{len(labels)}")

print(f"é…å°æˆåŠŸï¼š{len(dataset)} çµ„")

# åˆ†å‰²è³‡æ–™é›†
print("\nåˆ†å‰²è³‡æ–™é›†")

n_train, n_val, n_test = 9, 3, 3

# è‡ªå‹•èª¿æ•´æ•¸é‡
if len(dataset) < 15:
    n_test = min(3, len(dataset))
    n_val = min(3, len(dataset) - n_test)
    n_train = len(dataset) - n_test - n_val

# åˆ†å‰²ï¼šå…ˆåˆ†å‡ºæ¸¬è©¦é›†
remaining, test_files = train_test_split(dataset, test_size=n_test, random_state=42)

# å†å¾å‰©é¤˜åˆ†å‡ºé©—è­‰é›†
if len(remaining) > n_val:
    train_files, val_files = train_test_split(remaining, test_size=n_val, random_state=42)
else:
    train_files, val_files = [], remaining

# é™åˆ¶è¨“ç·´é›†æ•¸é‡
train_files = train_files[:n_train]

# ç”¢ç”Ÿ JSON
data_dict = {
    "training": train_files,
    "validation": val_files,
    "test": test_files
}

json_path = f"{base_path}/AICUP_training.json"
with open(json_path, "w", encoding='utf-8') as f:
    json.dump(data_dict, f, indent=2, ensure_ascii=False)

# é¡¯ç¤ºçµæœ
print("\n" + "=" * 70)
print("è³‡æ–™é›†æº–å‚™å®Œæˆï¼")
print("=" * 70)
print(f"è¨“ç·´é›†ï¼š{len(train_files)} å€‹æ¨£æœ¬")
print(f"é©—è­‰é›†ï¼š{len(val_files)} å€‹æ¨£æœ¬")
print(f"æ¸¬è©¦é›†ï¼š{len(test_files)} å€‹æ¨£æœ¬")
print(f"JSON è·¯å¾‘ï¼š{json_path}")

# é¡¯ç¤ºç¯„ä¾‹
print("\nè¨“ç·´é›†ç¯„ä¾‹ï¼š")
for i, item in enumerate(train_files[:3], 1):
    print(f"  {i}. {os.path.basename(item['image'])} â†” {os.path.basename(item['label'])}")

print("\næ‰€æœ‰è³‡æ–™æº–å‚™å®Œæˆï¼Œå¯ä»¥é–‹å§‹è¨“ç·´ï¼")
```

```python
# è·¯å¾‘è¨­å®šèˆ‡ JSON é©—è­‰

import glob
import random

# åŸºæœ¬è¨­å®š
workspace_dir = '/content/CardiacSegV2'
model_name = 'swinunetr'  # â­ æ”¹ç‚º swinunetr
data_name = 'chgh'
exp_name = 'AICUP_swinunetr_optuna'

# è·¯å¾‘è¨­å®š
root_data_dir = os.path.join(workspace_dir, 'dataset', data_name)
data_dicts_json = os.path.join(root_data_dir, 'AICUP_training.json')
root_exp_dir = os.path.join(workspace_dir, 'exps', 'exps', model_name, data_name, 'tune_results')

model_dir = os.path.join(workspace_dir, 'models')
log_dir = os.path.join(workspace_dir, 'logs')
eval_dir = os.path.join(workspace_dir, 'evals')

# å»ºç«‹æ‰€æœ‰å¿…è¦ç›®éŒ„
for dir_path in [model_dir, log_dir, eval_dir, root_exp_dir]:
    os.makedirs(dir_path, exist_ok=True)

# æ¨¡å‹æª”æ¡ˆè·¯å¾‘
best_checkpoint = os.path.join(model_dir, 'best_model.pth')
final_checkpoint = os.path.join(model_dir, 'final_model.pth')

print("\nè·¯å¾‘è¨­å®š:")
print(f"   å·¥ä½œç›®éŒ„: {workspace_dir}")
print(f"   æ¨¡å‹: {model_name}")
print(f"   JSON: {data_dicts_json}")
print(f"   æ¨¡å‹ç›®éŒ„: {model_dir}")

# æƒæä¸¦é…å°æª”æ¡ˆ
print("\næƒææª”æ¡ˆ...")

images = sorted(glob.glob(f"{root_data_dir}/training_image/*.nii.gz"))
labels = sorted(glob.glob(f"{root_data_dir}/training_label/*.nii.gz"))

print(f"   å½±åƒ: {len(images)} å€‹")
print(f"   æ¨™ç±¤: {len(labels)} å€‹")

# å»ºç«‹æ¨™ç±¤å¿«é€ŸæŸ¥æ‰¾å­—å…¸
label_dict = {}
for lbl in labels:
    base_name = os.path.basename(lbl).replace('_gt.nii.gz', '').replace('.nii.gz', '')
    label_dict[base_name] = lbl

# é…å°å½±åƒå’Œæ¨™ç±¤
dataset = []
for img in images:
    base_name = os.path.basename(img).replace('.nii.gz', '')
    if base_name in label_dict:
        dataset.append({"image": img, "label": label_dict[base_name]})

print(f"   é…å°æˆåŠŸ: {len(dataset)} çµ„")

# å›ºå®šåˆ†å‰² 9:3:3
if len(dataset) >= 15:
    print("\n åˆ†å‰²è³‡æ–™é›† (9:3:3)...")

    random.seed(42)
    dataset_shuffled = dataset.copy()
    random.shuffle(dataset_shuffled)

    train_files = dataset_shuffled[0:9]
    val_files = dataset_shuffled[9:12]
    test_files = dataset_shuffled[12:15]

    # å„²å­˜ JSON
    data_dict = {
        "train": train_files,
        "val": val_files,
        "test": test_files
    }

    with open(data_dicts_json, "w", encoding='utf-8') as f:
        json.dump(data_dict, f, indent=2, ensure_ascii=False)

    print(f"   è¨“ç·´é›†: {len(train_files)} å€‹")
    print(f"   é©—è­‰é›†: {len(val_files)} å€‹")
    print(f"   æ¸¬è©¦é›†: {len(test_files)} å€‹")
    print(f"   JSON å·²å„²å­˜")

elif len(dataset) > 0:
    print(f"\nè³‡æ–™ä¸è¶³ 15 çµ„ (åªæœ‰ {len(dataset)} çµ„)")
    print("   å°‡ä½¿ç”¨ç¾æœ‰è³‡æ–™é€²è¡Œåˆ†å‰²...")

    # è‡ªå‹•èª¿æ•´åˆ†å‰²æ¯”ä¾‹
    n_test = min(3, len(dataset) // 5)
    n_val = min(3, (len(dataset) - n_test) // 4)
    n_train = len(dataset) - n_test - n_val

    random.seed(42)
    dataset_shuffled = dataset.copy()
    random.shuffle(dataset_shuffled)

    train_files = dataset_shuffled[:n_train]
    val_files = dataset_shuffled[n_train:n_train+n_val]
    test_files = dataset_shuffled[n_train+n_val:]

    data_dict = {
        "train": train_files,
        "val": val_files,
        "test": test_files
    }

    with open(data_dicts_json, "w", encoding='utf-8') as f:
        json.dump(data_dict, f, indent=2, ensure_ascii=False)

    print(f"   è¨“ç·´é›†: {len(train_files)} å€‹")
    print(f"   é©—è­‰é›†: {len(val_files)} å€‹")
    print(f"   æ¸¬è©¦é›†: {len(test_files)} å€‹")

else:
    raise ValueError("ç„¡æ³•é…å°ä»»ä½•æª”æ¡ˆï¼è«‹æª¢æŸ¥è³‡æ–™è·¯å¾‘")

# é©—è­‰ JSON
print("\né©—è­‰ JSON...")

with open(data_dicts_json, 'r') as f:
    verify = json.load(f)

all_ok = True
for key in ["train", "val", "test"]:
    if key in verify and len(verify[key]) > 0:
        print(f"   âœ… {key}: {len(verify[key])} å€‹æ¨£æœ¬")

        # æª¢æŸ¥ç¬¬ä¸€ç­†æª”æ¡ˆæ˜¯å¦å­˜åœ¨
        first = verify[key][0]
        if not (os.path.exists(first["image"]) and os.path.exists(first["label"])):
            print(f"      âŒ æª”æ¡ˆä¸å­˜åœ¨ï¼")
            all_ok = False
    else:
        print(f"   âŒ {key}: è³‡æ–™ç¼ºå¤±")
        all_ok = False

print("\n" + "=" * 70)
if all_ok:
    print("è¨­å®šå®Œæˆï¼å¯ä»¥é–‹å§‹è¨“ç·´ï¼")
    print("\nè¨“ç·´æç¤º:")
    print("   - æ¨¡å‹æœƒå„²å­˜åœ¨:", model_dir)
    print("   - æ—¥èªŒæœƒå„²å­˜åœ¨:", log_dir)
    print("   - æœ€ä½³æ¨¡å‹:", best_checkpoint)
else:
    print("é©—è­‰å¤±æ•—ï¼è«‹æª¢æŸ¥ä¸Šæ–¹éŒ¯èª¤è¨Šæ¯")
print("=" * 70)
```

```python
# é‡æ–°åˆ†å‰²è³‡æ–™é›† (40:5:5)

import json
import glob
import os
from sklearn.model_selection import train_test_split

data_dir = "/content/CardiacSegV2/dataset/chgh"

# æƒææª”æ¡ˆ
images = sorted(glob.glob(f"{data_dir}/training_image/*.nii.gz"))
labels = sorted(glob.glob(f"{data_dir}/training_label/*_gt.nii.gz"))

# é…å°
dataset = []
for img in images:
    img_name = os.path.basename(img).replace('.nii.gz', '')
    label_name = f"{img_name}_gt.nii.gz"
    label_path = os.path.join(f"{data_dir}/training_label", label_name)
    if os.path.exists(label_path):
        dataset.append({"image": img, "label": label_path})

print(f"ç¸½å…± {len(dataset)} çµ„è³‡æ–™")

# åˆ†å‰² 40:5:5
n_train, n_val, n_test = 40, 5, 5

remaining, test_files = train_test_split(dataset, test_size=n_test, random_state=42)
train_files, val_files = train_test_split(remaining, test_size=n_val, random_state=42)

if len(train_files) > n_train:
    train_files = train_files[:n_train]

# å„²å­˜ JSON (æ³¨æ„æ˜¯ train å’Œ val)
data_dict = {
    "train": train_files,
    "val": val_files,
    "test": test_files
}

json_path = f"{data_dir}/AICUP_training.json"
with open(json_path, "w", encoding='utf-8') as f:
    json.dump(data_dict, f, indent=2, ensure_ascii=False)

print(f"\né‡æ–°åˆ†å‰²å®Œæˆ")
print(f"   è¨“ç·´é›†: {len(train_files)} å€‹")
print(f"   é©—è­‰é›†: {len(val_files)} å€‹")
print(f"   æ¸¬è©¦é›†: {len(test_files)} å€‹")
```
